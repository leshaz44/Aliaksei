Напишите функцию, которая принимает строку и возвращает True, если строка является палиндромом (игнорируя регистр и пробелы), и False в противном случае.

```python
def is_palindrome(s):
    # Удаляем пробелы и приводим строку к нижнему регистру
    processed_s = "".join(filter(str.isalnum, s)).lower()
    # Сравниваем обработанную строку с ее перевернутой версией
    return processed_s == processed_s[::-1]

# Получаем ввод от пользователя
user_input = input("Введите строку для проверки: ")

# Проверяем и выводим результат
if is_palindrome(user_input):
    print("Введенная строка является палиндромом!")
else:
    print("Введенная строка не является палиндромом.")

# Пример использования с конкретной строкой
print("\nПроверка примера 'Madam':", is_palindrome("Madam"))
```
