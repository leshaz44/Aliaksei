```python
# main.py
import ml_module as ml
import data_loader as dt
from sklearn.model_selection import train_test_split
import visualization as graf
from data_processing import count_missing_values

# Загрузка данных
file_path = 'data_Salary2.csv'  # Замените на путь к вашему CSV файлу
data = dt.load_data(file_path)

# Предобработка данных
target_column = 'Salary'
X, y, preprocessor = ml.preprocess_data(data, target_column)

# Подсчет пустых значений в каждом столбце
missing_values = count_missing_values(data)
print(missing_values)

# Разделение данных на тренировочную и тестовую выборки
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Обучение модели
model = ml.train_model(X_train, y_train)

# Предсказание на тестовых данных
y_pred = ml.predict(model, X_test)

# Оценка модели
mse, r2 = ml.evaluate_model(y_test, y_pred)
print(f"Среднеквадратичная ошибка: {mse:.2f}")
print(f"Коэффициент детерминации R^2: {r2:.2f}")

#visualization.py
# Визуализация истинных и предсказанных значений
graf.plot_predictions(y_test, y_pred)
# Визуализация гистограммы
graf.plot_histogram_errors(y_test, y_pred)
# Визуализация диаграммы рассеяния
graf.create_scatter_plot(y_test, y_pred)

```
